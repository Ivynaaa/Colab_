```python
#Ivyna Magalhães
#Beatriz Leonara

import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

imagem = cv2.imread('Lenna.jpg')

#coordenadas dos olhos
coordenadas_olho = (220, 250)

#largura e a altura da tarja preta
largura_tarja = 150
altura_tarja = 30

imagem[coordenadas_olho[1]:coordenadas_olho[1] + altura_tarja,
       coordenadas_olho[0]:coordenadas_olho[0] + largura_tarja] = 0  # 0 cor preta

cv2_imshow(imagem)
```

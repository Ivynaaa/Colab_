```python
#Beatriz Leonara
#Ivyna Magalhães

import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

lena_img = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)

#filtro de sobel
sobel_x = cv2.Sobel(lena_img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(lena_img, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobel_x, sobel_y)

#filtro de Canny
canny = cv2.Canny(lena_img, 50, 150)


print('Filtro de sobel')
cv2_imshow(sobel)
print('\n\n')
print('Filtro de canny')
cv2_imshow(canny)
```
